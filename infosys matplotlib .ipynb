```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)

# Plotting the data
plt.plot(x, y, 'co-', label='y = cos(2πx) * exp(-x)')  

# Adding grid lines
plt.grid(True)

# Adding labels and title
plt.xlabel('Abscissa')
plt.ylabel('Ordinate')

# Adding annotation
plt.annotate('θ = 60°', xy=(0.75, 0.25), xytext=(1.5, 0.5))
plt.annotate('2nd crest', xy=(2.25, 0.1), xytext=(2.75, 0.3), arrowprops=dict(facecolor='green', shrink=0.05))

# Setting the limits for y-axis
plt.ylim(-0.75, 1.0)

# Show plot
plt.show()
```
