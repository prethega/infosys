```python
import pandas as pd
data=pd.read_csv("D:/subashini/dataset/rainfall.csv")
data
```

```python
#missing values
```

```python
print(data.isnull().sum())
data = data.dropna()
data
```

```python
#district with highest annual rainfall
```

```python
data['Annual'] = data.loc[:, 'JAN':'DEC'].sum(axis=1)
max_rainfall = data.loc[data['Annual'].idxmax(), 'DISTRICT']
print("District with the highest annual rainfall:", max_rainfall)
```

```python
#top5 states with highest annual rainfall
```

```python
top = data.groupby('STATE_UT_NAME')['Annual'].sum().nlargest(5)
print("Top 5 states with the highest annual rainfall:\n", top)
```

```python
#Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'
```

```python
columns_to_drop = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']
data = data.drop(columns=columns_to_drop, errors='ignore')
data
```

```python
#Display the state-wise mean rainfall for all the months using a pivot table
```

```python
pivot_table = data.pivot_table(index='STATE_UT_NAME', values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], aggfunc='mean')
print("State-wise mean rainfall for all months:\n", pivot_table)
```

```python
#Display the count of districts in each state
```

```python
district_count = data['STATE_UT_NAME'].value_counts()
print("Count of districts in each state:\n", district_count)
```

```python
#For each state, display the district that gets the highest rainfall in May and the recorded rainfall
```

```python
highest_may_rainfall = data.loc[data.groupby('STATE_UT_NAME')['MAY'].idxmax()][['STATE_UT_NAME', 'DISTRICT', 'MAY']]
highest_may_rainfall
```
